# Tugas: Normalisasi Data Min-Max & WKNN (Weighted K-Nearest Neighbor)

---

## 📋 Soal

Diketahui data mahasiswa sebagai berikut:

| NO | IPK | PO (Uang Saku) | JMLH (Jumlah SKS) |
|----|-----|----------------|--------------------|
| 1  | 2   | 200.000        | 2                  |
| 2  | 3   | 300.000        | 3                  |
| 3  | 4   | 200.000        | 2                  |
| 4  | 2   | 200.000        | 3                  |
| 5  | 3   | 300.000        | 2                  |
| 6  | 4   | 400.000        | 3                  |
| 7  | 2   | 300.000        | **?**              |

**Pertanyaan:**
1. Lakukan **Normalisasi Min-Max** pada setiap fitur (IPK, PO, JMLH).
2. Gunakan metode **WKNN (Weighted K-Nearest Neighbor)** dengan **k=3** untuk memprediksi nilai **JMLH** pada data NO=7.

> **Keterangan:**  
> - IPK = Indeks Prestasi Kumulatif  
> - PO  = Pendapatan / Uang saku per bulan  
> - JMLH = Jumlah mata kuliah yang diambil  

---

## ✍️ JAWABAN MANUAL

### Langkah 1: Normalisasi Min-Max

**Rumus Normalisasi Min-Max:**

$$x' = \frac{x - x_{min}}{x_{max} - x_{min}}$$

---

#### Nilai Min dan Max setiap fitur:

| Fitur | Min     | Max     |
|-------|---------|---------|
| IPK   | 2       | 4       |
| PO    | 200.000 | 400.000 |
| JMLH  | 2       | 3       |

---

#### Perhitungan Normalisasi IPK:

$$IPK'_1 = \frac{2-2}{4-2} = \frac{0}{2} = 0$$

$$IPK'_2 = \frac{3-2}{4-2} = \frac{1}{2} = 0{,}5$$

$$IPK'_3 = \frac{4-2}{4-2} = \frac{2}{2} = 1$$

$$IPK'_4 = \frac{2-2}{4-2} = 0 \quad IPK'_5 = 0{,}5 \quad IPK'_6 = 1 \quad IPK'_7 = 0$$

---

#### Perhitungan Normalisasi PO:

$$PO'_1 = \frac{200000-200000}{400000-200000} = \frac{0}{200000} = 0$$

$$PO'_2 = \frac{300000-200000}{400000-200000} = \frac{100000}{200000} = 0{,}5$$

$$PO'_3 = \frac{200000-200000}{400000-200000} = 0 \quad PO'_4 = 0 \quad PO'_5 = 0{,}5 \quad PO'_6 = 1 \quad PO'_7 = 0{,}5$$

---

#### Perhitungan Normalisasi JMLH:

$$JMLH'_1 = \frac{2-2}{3-2} = 0 \quad JMLH'_2 = \frac{3-2}{3-2} = 1 \quad JMLH'_3 = 0$$

$$JMLH'_4 = 1 \quad JMLH'_5 = 0 \quad JMLH'_6 = 1 \quad JMLH'_7 = ?$$

---

#### Tabel Hasil Normalisasi:

| NO | IPK' | PO'  | JMLH' |
|----|------|------|-------|
| 1  | 0    | 0    | 0     |
| 2  | 0,5  | 0,5  | 1     |
| 3  | 1    | 0    | 0     |
| 4  | 0    | 0    | 1     |
| 5  | 0,5  | 0,5  | 0     |
| 6  | 1    | 1    | 1     |
| 7  | 0    | 0,5  | ?     |

---

### Langkah 2: Menghitung Jarak Euclidean dari Data NO=7 ke semua data latih

**Query point (NO=7):** IPK'=0, PO'=0,5  
**Rumus Euclidean Distance:**

$$d(a,b) = \sqrt{\sum_{i=1}^{n}(a_i - b_i)^2}$$

---

$$d(7,1) = \sqrt{(0-0)^2 + (0{,}5-0)^2} = \sqrt{0 + 0{,}25} = \sqrt{0{,}25} = 0{,}5$$

$$d(7,2) = \sqrt{(0-0{,}5)^2 + (0{,}5-0{,}5)^2} = \sqrt{0{,}25 + 0} = \sqrt{0{,}25} = 0{,}5$$

$$d(7,3) = \sqrt{(0-1)^2 + (0{,}5-0)^2} = \sqrt{1 + 0{,}25} = \sqrt{1{,}25} \approx 1{,}118$$

$$d(7,4) = \sqrt{(0-0)^2 + (0{,}5-0)^2} = \sqrt{0 + 0{,}25} = \sqrt{0{,}25} = 0{,}5$$

$$d(7,5) = \sqrt{(0-0{,}5)^2 + (0{,}5-0{,}5)^2} = \sqrt{0{,}25 + 0} = \sqrt{0{,}25} = 0{,}5$$

$$d(7,6) = \sqrt{(0-1)^2 + (0{,}5-1)^2} = \sqrt{1 + 0{,}25} = \sqrt{1{,}25} \approx 1{,}118$$

---

#### Tabel Jarak (diurutkan dari terkecil):

| NO | Jarak  | JMLH |
|----|--------|------|
| 1  | 0,5    | 2    |
| 2  | 0,5    | 3    |
| 4  | 0,5    | 3    | ← k=3 (3 tetangga terdekat)
| 5  | 0,5    | 2    |
| 3  | 1,118  | 2    |
| 6  | 1,118  | 3    |

**3 tetangga terdekat (k=3):** NO=1, NO=2, NO=4

---

### Langkah 3: Menghitung Bobot WKNN

**Rumus Bobot WKNN:**

$$w_i = \frac{1}{d_i^2}$$

$$w_1 = \frac{1}{0{,}5^2} = \frac{1}{0{,}25} = 4$$

$$w_2 = \frac{1}{0{,}5^2} = \frac{1}{0{,}25} = 4$$

$$w_4 = \frac{1}{0{,}5^2} = \frac{1}{0{,}25} = 4$$

---

### Langkah 4: Voting Berbobot (Weighted Voting)

| Kelas (JMLH) | Tetangga | Bobot | Total Bobot |
|--------------|----------|-------|-------------|
| JMLH = 2     | NO=1     | 4     | **4**       |
| JMLH = 3     | NO=2, NO=4 | 4+4 | **8**       |

$$\text{Total bobot JMLH=2} = 4$$
$$\text{Total bobot JMLH=3} = 4 + 4 = 8$$

### ✅ Kesimpulan Manual:

> Kelas dengan bobot tertinggi adalah **JMLH = 3** (bobot = 8)  
> Maka prediksi untuk data NO=7 adalah $\boxed{JMLH = 3}$

## 💻 JAWABAN CODING (Python)

```python
import numpy as np
import pandas as pd
from collections import defaultdict

# ============================================================
# DATA AWAL
# ============================================================
data = {
    'NO'  : [1, 2, 3, 4, 5, 6, 7],
    'IPK' : [2, 3, 4, 2, 3, 4, 2],
    'PO'  : [200000, 300000, 200000, 200000, 300000, 400000, 300000],
    'JMLH': [2, 3, 2, 3, 2, 3, np.nan]  # NO=7 belum diketahui
}

df = pd.DataFrame(data)
print("="*55)
print("           DATA AWAL")
print("="*55)
print(df.to_string(index=False))
print("="*55)
```

```python
# ============================================================
# LANGKAH 1: NORMALISASI MIN-MAX
# ============================================================

def min_max_normalize(series):
    """Normalisasi Min-Max: x' = (x - x_min) / (x_max - x_min)"""
    x_min = series.min()
    x_max = series.max()
    return (series - x_min) / (x_max - x_min), x_min, x_max

df_norm = df.copy()

# Normalisasi IPK (gunakan semua data termasuk NO=7)
df_norm['IPK_norm'], ipk_min, ipk_max = min_max_normalize(df['IPK'])

# Normalisasi PO (gunakan semua data termasuk NO=7)
df_norm['PO_norm'], po_min, po_max = min_max_normalize(df['PO'])

# Normalisasi JMLH (hanya data latih, NO=7 diabaikan)
jmlh_train = df['JMLH'].dropna()
jmlh_min = jmlh_train.min()
jmlh_max = jmlh_train.max()
df_norm['JMLH_norm'] = (df['JMLH'] - jmlh_min) / (jmlh_max - jmlh_min)

print("="*65)
print("           TABEL NORMALISASI MIN-MAX")
print("="*65)
print(f"  Min-Max IPK  : min={ipk_min}, max={ipk_max}")
print(f"  Min-Max PO   : min={po_min:,}, max={po_max:,}")
print(f"  Min-Max JMLH : min={jmlh_min}, max={jmlh_max}")
print("-"*65)

display_cols = ['NO', 'IPK', 'PO', 'JMLH', 'IPK_norm', 'PO_norm', 'JMLH_norm']
print(df_norm[display_cols].to_string(index=False))
print("="*65)
```

```python
# ============================================================
# LANGKAH 2: MENGHITUNG JARAK EUCLIDEAN
# ============================================================

def euclidean_distance(point_a, point_b):
    """Hitung jarak Euclidean antara dua titik"""
    return np.sqrt(np.sum((np.array(point_a) - np.array(point_b))**2))

# Data latih (NO=1 sampai NO=6)
train_data = df_norm[df_norm['JMLH'].notna()].copy()

# Data uji / query point (NO=7)
query = df_norm[df_norm['NO'] == 7][['IPK_norm', 'PO_norm']].values[0]

print("="*55)
print("  QUERY POINT (NO=7):")
print(f"  IPK_norm = {query[0]}, PO_norm = {query[1]}")
print("="*55)
print("  PERHITUNGAN JARAK EUCLIDEAN:")
print("-"*55)

distances = []
for _, row in train_data.iterrows():
    point = [row['IPK_norm'], row['PO_norm']]
    dist  = euclidean_distance(query, point)
    distances.append({
        'NO'   : int(row['NO']),
        'IPK_norm': row['IPK_norm'],
        'PO_norm' : row['PO_norm'],
        'JMLH'    : int(row['JMLH']),
        'Jarak'   : round(dist, 4)
    })
    diff_ipk = f"({query[0]}-{row['IPK_norm']})²"
    diff_po  = f"({query[1]}-{row['PO_norm']})²"
    print(f"  d(7,{int(row['NO'])}) = sqrt{diff_ipk} + {diff_po} = {dist:.4f}")

df_dist = pd.DataFrame(distances).sort_values('Jarak').reset_index(drop=True)

print("\n" + "="*55)
print("  TABEL JARAK (urut dari terkecil):")
print("="*55)
print(df_dist.to_string(index=False))
print("="*55)
```

```python
# ============================================================
# LANGKAH 3: WKNN - PILIH k=3 TETANGGA TERDEKAT & HITUNG BOBOT
# ============================================================

k = 3
k_nearest = df_dist.head(k).copy()

# Bobot WKNN: w = 1 / d²
k_nearest['Bobot (1/d²)'] = 1 / (k_nearest['Jarak'] ** 2)

print("="*60)
print(f"  {k} TETANGGA TERDEKAT (k={k}) & BOBOT WKNN:")
print("="*60)
print(k_nearest[['NO', 'IPK_norm', 'PO_norm', 'Jarak', 'JMLH', 'Bobot (1/d²)']].to_string(index=False))
print("="*60)
```

```python
# ============================================================
# LANGKAH 4: WEIGHTED VOTING - PREDIKSI KELAS
# ============================================================

# Akumulasi bobot per kelas JMLH
class_weights = defaultdict(float)
for _, row in k_nearest.iterrows():
    class_weights[int(row['JMLH'])] += row['Bobot (1/d²)']

print("="*55)
print("  WEIGHTED VOTING:")
print("="*55)
for kelas, bobot in sorted(class_weights.items()):
    detail = k_nearest[k_nearest['JMLH'] == kelas]
    no_list = list(detail['NO'].astype(int))
    print(f"  JMLH = {kelas}  |  Tetangga: NO={no_list}  |  Total Bobot = {bobot:.4f}")

# Prediksi: kelas dengan total bobot tertinggi
prediksi_jmlh = max(class_weights, key=class_weights.get)

print("="*55)
print(f"  ✅ HASIL PREDIKSI WKNN untuk NO=7:")
print(f"     JMLH = {prediksi_jmlh}")
print("="*55)
```

```python
# ============================================================
# REKAP LENGKAP - TABEL AKHIR
# ============================================================

df_final = df_norm[['NO', 'IPK', 'PO', 'JMLH', 'IPK_norm', 'PO_norm', 'JMLH_norm']].copy()

# Isi prediksi untuk NO=7
df_final.loc[df_final['NO'] == 7, 'JMLH'] = prediksi_jmlh
jmlh_norm_pred = (prediksi_jmlh - jmlh_min) / (jmlh_max - jmlh_min)
df_final.loc[df_final['NO'] == 7, 'JMLH_norm'] = jmlh_norm_pred

print("="*75)
print("  TABEL LENGKAP: DATA AWAL + NORMALISASI + PREDIKSI WKNN")
print("="*75)
print(df_final.to_string(index=False))
print("="*75)
print(f"  Catatan: Nilai JMLH pada NO=7 diprediksi dengan WKNN (k={k}) = {prediksi_jmlh}")
print("="*75)
```
